```python
import os
actors = os.listdir('TrainingData')

filenames =[]
for actor in actors:
    for file in os.listdir(os.path.join('TrainingData',actor)):
        filenames.append(os.path.join('TrainingData',actor,file)) 
```

```python
print(filenames)
print(len(filenames))
```

```python
import pickle
pickle.dump(filenames,open('filenames.pkl','wb'))
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image
from keras_vggface.utils import preprocess_input
from keras_vggface.vggface import VGGFace
```

```python
filenames=pickle.load(open('filenames.pkl','rb'))
model = VGGFace(model='resnet50',include_top=False,input_shape=(244,244,3),pooling='avg')
```

```python
print(model.summary())
```

```python
def featureExtractor(imgPath,model):
    img = image.load_img(imgPath,target_size=(224,224))
    imgArray = image.img_to_array(img)
    expandImage = np.expand_dims(imgArray,axis=0)
    preprocessedImage = preprocess_input(expandImage)
    
    result = model.predict(preprocessedImage).flatten()
    
    return result
```

```python
from tqdm import tqdm
features = []
for file in tqdm(filenames):
    features.append(featureExtractor(file,model))

pickle.dump(features,open('embedding.plk','wb'))
```
